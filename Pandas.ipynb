```python
import pandas as pd
from sklearn.datasets import fetch_california_housing

# Load California dataset
california = fetch_california_housing()
california_df = pd.DataFrame(california.data, columns=california.feature_names)
california_df['Price'] = california.target  # This works now

# Load Boston dataset
boston_url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_df = pd.read_csv(boston_url)

# Check first few rows
print("Boston Dataset:")
print(boston_df.head())
```

```python
#removing particular row
# for particular removal to row, axis = 0
# for particular removal to column, axis = 1
boston_df.drop(index=0, axis=0)
```

```python
#removing particular column
boston_df.drop(columns='zn', axis=1)
```
